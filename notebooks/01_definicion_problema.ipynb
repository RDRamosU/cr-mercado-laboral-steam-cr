## 1. Contexto

Costa Rica ha consolidado en las últimas décadas un ecosistema tech
robusto, impulsado principalmente por empresas multinacionales instaladas
bajo el Régimen de Zona Franca y un sector exportador de servicios TIC
en crecimiento acelerado.

Este crecimiento genera una pregunta fundamental: ¿el sistema educativo
costarricense produce suficientes graduados STEAM para alimentar esa
demanda? O dicho de otra forma ¿existe una brecha entre la oferta de
talento tech y los empleos disponibles?

Este proyecto analiza esa brecha utilizando datos oficiales del CONARE,
MICITT, BCCR, PROCOMER e INEC, cubriendo el periodo 2014–2024.

**Nota de continuidad:** Este proyecto es la continuación directa del
Proyecto 1 (Graduados STEAM en CR 2014–2022), donde se estableció que
las universidades estatales producen entre 5.169 y 7.273 graduados
STEAM anuales. Aquí analizamos si ese volumen es suficiente para cubrir
la demanda del mercado laboral tech.

## 2. Pregunta central

> ¿Existe una brecha entre los graduados STEAM que produce Costa Rica
> y los empleos tech disponibles?

## 3. Preguntas de análisis

1. ¿Cuál es el desempleo de graduados STEM vs el promedio nacional
   en 2019 y 2022?
2. ¿A qué ritmo crece el empleo en el sector TIC vs la producción
   de graduados STEAM?
3. ¿Qué disciplinas tech tienen mayor inserción laboral y menor
   desvinculación con la carrera?
4. ¿Existe una brecha de género en la inserción laboral STEM?
5. ¿Cómo se comparan los salarios del sector TIC vs el promedio nacional?

## 4. Hallazgo anticipado

A diferencia de lo que podría esperarse, el análisis exploratorio de
los datos revela que:

> **La brecha de talento tech en Costa Rica no es de exceso de graduados
> sin empleo — es exactamente la contraria.**

El mercado absorbe prácticamente todos los graduados tech disponibles
y crece más rápido que la capacidad del sistema educativo de producirlos.
